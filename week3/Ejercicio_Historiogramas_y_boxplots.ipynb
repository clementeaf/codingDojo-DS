```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/My Drive/catsvdogs.xlsx'
df = pd.read_excel(filename)
df
```

###  Trazar un histograma del porcentaje de dueños de perros.

```python
# Extraemos la columna de porcentaje de dueños de perros
porcentaje_dog_owners = df['Percentage of Dog Owners']

# Creamos el histograma
plt.hist(porcentaje_dog_owners, bins=10, edgecolor='black')

# Agregamos etiquetas y título
plt.xlabel('Porcentaje de dueños de perros')
plt.ylabel('Frecuencia')
plt.title('Histograma del Porcentaje de Dueños de Perros')

# Mostramos el gráfico
plt.show()
```

## Tracen un histograma del porcentaje de dueños de gatos.

```python
# Extraemos la columna de porcentaje de dueños de gatos
porcentaje_cat_owners = df['Percentage of Cat Owners']

# Creamos el histograma
plt.hist(porcentaje_cat_owners, bins=10, edgecolor='black')

# Agregamos etiquetas y título
plt.xlabel('Porcentaje de dueños de gatos')
plt.ylabel('Frecuencia')
plt.title('Histograma del Porcentaje de Dueños de Gatos')

# Mostramos el gráfico
plt.show()
```

```python

# Extraemos las columnas de porcentaje de dueños de perros y gatos
porcentaje_dog_owners = df['Percentage of Dog Owners']
porcentaje_cat_owners = df['Percentage of Cat Owners']

# Creamos el histograma de dueños de perros
plt.hist(porcentaje_dog_owners, bins=10, edgecolor='black', alpha=0.5, label='Dueños de Perros')

# Creamos el histograma de dueños de gatos
plt.hist(porcentaje_cat_owners, bins=10, edgecolor='black', alpha=0.5, label='Dueños de Gatos')

# Agregamos etiquetas y título
plt.xlabel('Porcentaje de dueños')
plt.ylabel('Frecuencia')
plt.title('Comparación del Porcentaje de Dueños de Perros y Gatos')
plt.legend()

# Mostramos el gráfico
plt.show()

# A partir del primer historiograma

# Podemos ver la distribución de los porcentajes de dueños de gatos en el conjunto de datos.
# Por ejemplo, si hay más hogares con un porcentaje de dueños de gatos entre ciertos rangos,
# el histograma mostrará un pico o una mayor frecuencia en esos rangos.

# Podemos observar la forma general de la distribución, si es simétrica,
# sesgada hacia la izquierda o hacia la derecha, o si tiene múltiples picos.

# A partir del segundo historiograma
# La leyenda nos indica qué color corresponde a cada histograma (dueños de perros o dueños de gatos).
# Al comparar los dos histogramas, podemos ver la distribución de ambas variables y cómo se comparan entre sí.

# Si hay solapamiento entre los histogramas, significa que hay porcentajes similares de dueños de perros y
# gatos en el conjunto de datos. Si los histogramas tienen formas diferentes,
# puede indicar diferencias en las preferencias de tenencia de mascotas entre gatos y perros.
```

```python
# Generar dos boxplots:
# Uno con la media del número de gatos por hogar y el otro de la media del número de perros

# Creamos una nueva columna llamada 'Mascota' con los datos de los gatos
df['Mascota'] = df['Mean Number of Cats']

# Creamos una nueva columna llamada 'Tipo' con la etiqueta 'Gatos' para distinguir los gatos
df['Tipo'] = 'Gatos'

# Creamos un nuevo DataFrame para los datos de los perros
df_dogs = df.copy()

# Actualizamos la columna 'Mascota' con los datos de los perros
df_dogs['Mascota'] = df_dogs['Mean Number of Dogs per household']

# Actualizamos la columna 'Tipo' con la etiqueta 'Perros' para distinguir los perros
df_dogs['Tipo'] = 'Perros'

# Combinamos los DataFrames de gatos y perros utilizando pd.concat()
df_combined = pd.concat([df, df_dogs], ignore_index=True)

# Creamos el boxplot con los datos combinados
sns.boxplot(x='Tipo', y='Mascota', data=df_combined, palette=['skyblue', 'salmon'])

# Agregamos etiquetas y título
plt.xlabel('Mascotas')
plt.ylabel('Número Promedio por Hogar')
plt.title('Boxplots del Número Promedio de Gatos y Perros por Hogar')

# Mostramos el gráfico
plt.show()

# Al comparar los dos boxplots, podemos visualizar y comparar la distribución del número promedio
# de gatos y perros por hogar en el conjunto de datos.

# Podemos ver que la mediana del número promedio de gatos por hogar es más alta que la mediana del
# número promedio de perros por hogar. Esto sugiere que en promedio, hay más gatos por hogar que perros en
# el conjunto de datos.

# La variabilidad en el número promedio de gatos por hogar (longitud de la caja) es mayor que la variabilidad
# en el número promedio de perros por hogar. Esto indica que hay más variabilidad en la cantidad de gatos que
# en la cantidad de perros por hogar.

# Observamos puntos fuera de los bigotes en ambos boxplots, lo que indica la presencia de valores atípicos en
# ambas distribuciones. Estos valores atípicos pueden representar hogares con un número inusualmente alto de
# gatos o perros.
```
