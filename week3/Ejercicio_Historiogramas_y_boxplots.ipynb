```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/My Drive/catsvdogs.xlsx'
df = pd.read_excel(filename)
df
```

###  Trazar un histograma del porcentaje de dueos de perros.

```python
# Extraemos la columna de porcentaje de dueos de perros
porcentaje_dog_owners = df['Percentage of Dog Owners']

# Creamos el histograma
plt.hist(porcentaje_dog_owners, bins=10, edgecolor='black')

# Agregamos etiquetas y ttulo
plt.xlabel('Porcentaje de dueos de perros')
plt.ylabel('Frecuencia')
plt.title('Histograma del Porcentaje de Dueos de Perros')

# Mostramos el grfico
plt.show()
```

## Tracen un histograma del porcentaje de dueos de gatos.

```python
# Extraemos la columna de porcentaje de dueos de gatos
porcentaje_cat_owners = df['Percentage of Cat Owners']

# Creamos el histograma
plt.hist(porcentaje_cat_owners, bins=10, edgecolor='black')

# Agregamos etiquetas y ttulo
plt.xlabel('Porcentaje de dueos de gatos')
plt.ylabel('Frecuencia')
plt.title('Histograma del Porcentaje de Dueos de Gatos')

# Mostramos el grfico
plt.show()
```

```python

# Extraemos las columnas de porcentaje de dueos de perros y gatos
porcentaje_dog_owners = df['Percentage of Dog Owners']
porcentaje_cat_owners = df['Percentage of Cat Owners']

# Creamos el histograma de dueos de perros
plt.hist(porcentaje_dog_owners, bins=10, edgecolor='black', alpha=0.5, label='Dueos de Perros')

# Creamos el histograma de dueos de gatos
plt.hist(porcentaje_cat_owners, bins=10, edgecolor='black', alpha=0.5, label='Dueos de Gatos')

# Agregamos etiquetas y ttulo
plt.xlabel('Porcentaje de dueos')
plt.ylabel('Frecuencia')
plt.title('Comparacin del Porcentaje de Dueos de Perros y Gatos')
plt.legend()

# Mostramos el grfico
plt.show()

# A partir del primer historiograma

# Podemos ver la distribucin de los porcentajes de dueos de gatos en el conjunto de datos.
# Por ejemplo, si hay ms hogares con un porcentaje de dueos de gatos entre ciertos rangos,
# el histograma mostrar un pico o una mayor frecuencia en esos rangos.

# Podemos observar la forma general de la distribucin, si es simtrica,
# sesgada hacia la izquierda o hacia la derecha, o si tiene mltiples picos.

# A partir del segundo historiograma
# La leyenda nos indica qu color corresponde a cada histograma (dueos de perros o dueos de gatos).
# Al comparar los dos histogramas, podemos ver la distribucin de ambas variables y cmo se comparan entre s.

# Si hay solapamiento entre los histogramas, significa que hay porcentajes similares de dueos de perros y
# gatos en el conjunto de datos. Si los histogramas tienen formas diferentes,
# puede indicar diferencias en las preferencias de tenencia de mascotas entre gatos y perros.
```

```python
# Generar dos boxplots:
# Uno con la media del nmero de gatos por hogar y el otro de la media del nmero de perros

# Creamos una nueva columna llamada 'Mascota' con los datos de los gatos
df['Mascota'] = df['Mean Number of Cats']

# Creamos una nueva columna llamada 'Tipo' con la etiqueta 'Gatos' para distinguir los gatos
df['Tipo'] = 'Gatos'

# Creamos un nuevo DataFrame para los datos de los perros
df_dogs = df.copy()

# Actualizamos la columna 'Mascota' con los datos de los perros
df_dogs['Mascota'] = df_dogs['Mean Number of Dogs per household']

# Actualizamos la columna 'Tipo' con la etiqueta 'Perros' para distinguir los perros
df_dogs['Tipo'] = 'Perros'

# Combinamos los DataFrames de gatos y perros utilizando pd.concat()
df_combined = pd.concat([df, df_dogs], ignore_index=True)

# Creamos el boxplot con los datos combinados
sns.boxplot(x='Tipo', y='Mascota', data=df_combined, palette=['skyblue', 'salmon'])

# Agregamos etiquetas y ttulo
plt.xlabel('Mascotas')
plt.ylabel('Nmero Promedio por Hogar')
plt.title('Boxplots del Nmero Promedio de Gatos y Perros por Hogar')

# Mostramos el grfico
plt.show()

# Al comparar los dos boxplots, podemos visualizar y comparar la distribucin del nmero promedio
# de gatos y perros por hogar en el conjunto de datos.

# Podemos ver que la mediana del nmero promedio de gatos por hogar es ms alta que la mediana del
# nmero promedio de perros por hogar. Esto sugiere que en promedio, hay ms gatos por hogar que perros en
# el conjunto de datos.

# La variabilidad en el nmero promedio de gatos por hogar (longitud de la caja) es mayor que la variabilidad
# en el nmero promedio de perros por hogar. Esto indica que hay ms variabilidad en la cantidad de gatos que
# en la cantidad de perros por hogar.

# Observamos puntos fuera de los bigotes en ambos boxplots, lo que indica la presencia de valores atpicos en
# ambas distribuciones. Estos valores atpicos pueden representar hogares con un nmero inusualmente alto de
# gatos o perros.
```
