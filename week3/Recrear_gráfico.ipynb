```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = '/content/drive/MyDrive/mortgages.csv'
df = pd.read_csv(file)
```

```python
# Crear un filtro para que el nombre de la hipoteca sea de “30 años”.  Usen len() para encontrar cuántas filas corresponden a 30 años.
# Cargar el DataFrame con los datos proporcionados

# Obtener el nombre de las columnas
nombres_columnas = df.columns

# Filtrar las filas donde el nombre de la hipoteca sea "30 Years"
df_filtro_30_anos = df[df['Mortgage Name'] == '30 Year']

# Obtener el número de filas correspondientes a "30 Years"
filas_30_anos = len(df_filtro_30_anos)

# Imprimir el número de filas
print("Número de filas correspondientes a '30 Years':", filas_30_anos)

# Imprimir el nuevo DataFrame con el filtro aplicado
print(df_filtro_30_anos)
```

```python
# Crear un filtro para una tasa de interés del 3 % (observen cómo está escrito en los datos).

# Filtrar las filas donde el nombre de la hipoteca sea "30 Years"
df_filtro_30_anos = df[df['Mortgage Name'] == '30 Year']

# Crear un filtro para una tasa de interés del 3%
filtro_tasa_interes_3 = df_filtro_30_anos['Interest Rate'] == 0.03

# Aplicar el filtro al DataFrame
df_filtro_tasa_interes_3 = df_filtro_30_anos[filtro_tasa_interes_3]

# Imprimir el nuevo DataFrame con ambos filtros aplicados
print(df_filtro_tasa_interes_3)
```

```python
# Definir el nuevo DataFrame basándose en los filtros de año e interés que han creado.
# Filtrar las filas donde el nombre de la hipoteca sea "30 Years"
df_filtro_30_anos = df[df['Mortgage Name'] == '30 Year']

# Crear un filtro para una tasa de interés del 3%
filtro_tasa_interes_3 = df_filtro_30_anos['Interest Rate'] == 0.03

# Aplicar ambos filtros al DataFrame original
df_nuevo = df.loc[(df['Mortgage Name'] == '30 Year') & (df['Interest Rate'] == 0.03)]

# Imprimir el nuevo DataFrame
print(df_nuevo)
```

```python
# filtro para la tasa de interés del 5 %
# Filtrar las filas donde el nombre de la hipoteca sea "30 Years"
df_filtro_30_anos = df[df['Mortgage Name'] == '30 Year']

# Crear un filtro para una tasa de interés del 3%
filtro_tasa_interes_3 = df_filtro_30_anos['Interest Rate'] == 0.03

# Crear un filtro para una tasa de interés del 5%
filtro_tasa_interes_5 = df_filtro_30_anos['Interest Rate'] == 0.05

# Aplicar ambos filtros al DataFrame original para obtener dos nuevos DataFrames
df_nuevo_tasa_interes_3 = df_filtro_30_anos[filtro_tasa_interes_3]
df_nuevo_tasa_interes_5 = df_filtro_30_anos[filtro_tasa_interes_5]

# Combinar ambos filtros con el operador lógico "&" para obtener el nuevo DataFrame
df_nuevo = df_filtro_30_anos[filtro_tasa_interes_3 | filtro_tasa_interes_5]

# Imprimir el nuevo DataFrame con los filtros aplicados
print(df_nuevo)
```

```python
#  Puesto que queremos graficar la suma acumulativa del interés pagado, podemos usar la función .cumsum().
# ¿Cuáles son los tres primeros valores en la salida para la suma acumulativa de la tasa del 3 %?
# Aplicar la función .cumsum() a la columna "Interest Paid" del DataFrame filtrado con tasa de interés del 3%

suma_acumulativa_tasa_3 = df_nuevo_tasa_interes_3['Interest Paid'].cumsum()

# Obtener los tres primeros valores de la suma acumulativa de la tasa del 3%
primeros_valores_tasa_3 = suma_acumulativa_tasa_3.head(3)

# Imprimir los tres primeros valores
print("Los tres primeros valores de la suma acumulativa para la tasa del 3% son:")
print(primeros_valores_tasa_3)
```

```python
# Definir el DataFrame "mortgage30_03" con las filas filtradas para una tasa de interés del 3%
mortgage30_03 = df[(df['Mortgage Name'] == '30 Year') & (df['Interest Rate'] == 0.03)]

# Definir el DataFrame "mortgage30_05" con las filas filtradas para una tasa de interés del 5%
mortgage30_05 = df[(df['Mortgage Name'] == '30 Year') & (df['Interest Rate'] == 0.05)]
```

```python
# Usen plt.plot(x, y, c=color) para trazar la línea negra (5 %).  Queremos los valores x para que sean ['Month'] y
# los valores y para que sean la suma acumulativa del Interest Paid (interés pagado).
# Podemos usar el código de la pregunta anterior para definir los valores y dentro de la función plt.plot().
# Pista: para el negro, c = 'k'
# Supongamos que los DataFrames se llaman mortgage30_03 y mortgage30_05

# Obtener los valores de "Months" y "Interest Paid" para el DataFrame mortgage30_03
x_30_03 = mortgage30_03['Month']
y_30_03 = mortgage30_03['Interest Paid'].cumsum()

# Obtener los valores de "Months" y "Interest Paid" para el DataFrame mortgage30_05
x_30_05 = mortgage30_05['Month']
y_30_05 = mortgage30_05['Interest Paid'].cumsum()

# Trazar la línea azul correspondiente a la tasa de interés del 3%
plt.plot(x_30_03, y_30_03, c='b', label='Tasa 0.03', linestyle='-', linewidth=2)

# Trazar la línea negra correspondiente a la tasa de interés del 5%
plt.plot(x_30_05, y_30_05, c='k', label='Tasa 0.05', linestyle='-', linewidth=2)

# Agregar etiquetas y título
plt.xlabel('Months')
plt.ylabel('Suma Acumulativa del Interés Pagado')
plt.title('Suma Acumulativa del Interés Pagado por Mes')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
