```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/My Drive/kingCountyHouseData.csv'
df = pd.read_csv(filename)
df.head()
```

```python
# Take a look at the price column
df['price'].head()
```

```python
# Make a default histogram of the price column
df['price'].hist()
```

```python
# Try to remove scientific notation
df['price'].hist()
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
# rotate your xticklabels
df['price'].hist()
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
# Change number of bins
df['price'].hist(bins = 30)
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
# Nos enfocaremos en un subconjunto de datos que solo recoge los precios de 300 000 o menos.
# El código de abajo también agrega un contorno negro alrededor cada intervalo estableciendo el color del borde en negro.

# visualizing a subset of the data, making more bins, and visualizing a subset of data
# this was just to explore the data
price_filter = df.loc[:, 'price'] <= 3000000
df.loc[price_filter, 'price'].hist(bins = 30,
                                   edgecolor='black')
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
# Adding seaborn style
plt.style.use('seaborn')
price_filter = df.loc[:, 'price'] <= 3000000
df.loc[price_filter, 'price'].hist(bins = 30,
                                   edgecolor='black')
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```
